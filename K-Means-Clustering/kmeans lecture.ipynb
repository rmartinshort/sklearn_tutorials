Clustering in an unsupervised sense   
Groups clusters based on distance  
Segmenting physcial groups  
Observations in each group should be similar  

Choose k. Assign cluster definitions at random and calculate centroids. Assign clusters based on distance from the center. Keep repeating until datapoints stop being assigned

A very large k is associated with over complexity in the model (basically overfitting)

We use the elbow method to choose k - based on when the SSE changes abrupty 

```python
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
#Generate some data
from sklearn.datasets import make_blobs
```

```python
#This allows us to make artificial data to play with
data = make_blobs(n_samples=200,n_features=4,centers=4,
                  cluster_std=1.8,
                 random_state=101)
```

```python
data[0] #This is 200 x 2 - 200 samples of two features
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
#Labels it believes to be true. If we are dealing with real data we 
#would be unable to validate
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))


#Color data by the algorithm choices
ax1.set_title("K Means")
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,
            cmap='rainbow')

ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],
            cmap='rainbow')
```

Kmeans does a good job here
